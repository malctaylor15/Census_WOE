```python
import pandas as pd 
import numpy as np 
import sklearn as sk 
import matplotlib.pyplot as plt 
```

```python
columns = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation"\
           , "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week"\
           , "native-country", "class" ]
```

```python
data1 = pd.read_csv("adult.data", names= columns)
```

```python
data1.head(50)
```

```python
data1.describe()
```

### Begin dealing with One Hot Encoded Variables 

```python
from collections import Counter
```

```python
one_hot_var_names = ["workclass", "education", "marital-status", "occupation", "relationship", "race", "native-country"]
```

```python
one_hot_vars = pd.get_dummies(data1[one_hot_var_names])
```

```python
one_hot_var_counts_before = {}
for orig_var_name in one_hot_var_names:
    one_hot_var_counts_before[orig_var_name] = sum([1 if orig_var_name in dummy_var_name else 0 \
                                             for dummy_var_name in one_hot_vars.columns ])
```

```python
one_hot_var_counts_before
```

```python
one_hot_vars.columns
```

```python
all_one_hot_vars_mean = one_hot_vars.mean(axis = 0)
pd.Series(all_one_hot_vars_mean).describe()
```

Notice there are over 100 unique fields with most of them being less than 1.5% full..  
Let's keep the ones with more than 10% coverage. It should be around 70% or ~30 one hot encoded variables 

```python
min_pct_good = 0.10
good_one_hot_var_names = all_one_hot_vars_mean[all_one_hot_vars_mean > min_pct_good].index
print("Number of good vars: ", len(good_one_hot_var_names))
good_one_hot_var_names
```

```python
good_one_hot_vars = one_hot_vars[good_one_hot_var_names]
```

```python
one_hot_var_counts_after = {}
for orig_var_name in one_hot_var_names:
    one_hot_var_counts_after[orig_var_name] = sum([1 if orig_var_name in dummy_var_name else 0 \
                                             for dummy_var_name in good_one_hot_vars.columns ])
```

```python
# Number of variables before and after 
pd.DataFrame([one_hot_var_counts_before,one_hot_var_counts_after], index = ["before", "after"]).T
```

### Combine the one hots 

```python
data2 = data1.drop(one_hot_var_names, axis = 1)
data2 = pd.concat([data2, good_one_hot_vars], axis = 1)
# Fix the binary cats 
```

```python
print("Unique sex: ", data2["sex"].unique())
print("Unique class: ", data2["class"].unique())
```

```python
data2["sex"] = data2["sex"].apply(lambda x: 1 if x == " Male" else 0)
data2["class"] = data2["class"].apply(lambda x: 1 if x == " <=50K" else 0)
```

### Final preps for algorithms

```python
from sklearn.model_selection import train_test_split
```

```python
data3 = data2
data3["intercept"] = 1.0
```

```python
X = data3.drop(["class"], axis = 1)
Y = data3["class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

### Logistic Regression

```python
import statsmodels.api as sm 
```

```python
logit = sm.Logit(y_train, X_train)
```

```python
results = logit.fit()
```
